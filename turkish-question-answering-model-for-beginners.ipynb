# Turkish Question Answering Model for Beginners

This notebook provides a step-by-step guide for creating a simple question-answering model in Turkish using the Hugging Face Transformers library. The notebook demonstrates how to install the necessary libraries, load a pre-trained model and tokenizer, create a pipeline, and use the pipeline to answer questions based on a given context.

# Step 2: Import Libraries
Import the necessary classes and functions from the transformers library.

```python
from transformers import AutoModelForQuestionAnswering, AutoTokenizer, pipeline
```

# Step 3: Download Model and Tokenizer
Download the pre-trained BERT model and tokenizer for Turkish question answering.

```python
# Download the model and tokenizer
model_name = "savasy/bert-base-turkish-squad"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
```

Step 4: Create the Pipeline
Create a question-answering pipeline using the downloaded model and tokenizer.

```python
# Create the pipeline
nlp = pipeline("question-answering", model=model, tokenizer=tokenizer)
```

# Step 6: Define Questions
List the questions you want the model to answer based on the provided context.

```python
# Example questions
questions = [
"Mustafa Kemal Atatürk nerede doğdu?",
"Atatürk'ün babası kimdir?",
"Atatürk hangi tarihte vefat etti?",
"Atatürk'ün ilk eşi kimdir?",
"Atatürk'ün katıldığı savaşlar hangileridir?"
]
```

# Step 7: Get Answers from the Model
Use the pipeline to find answers to the questions based on the context.

```python
# Find answers using the model
for question in questions:
    result = nlp(question=question, context=context)
    print(f"Question: {question}")
    print(f"Answer: {result['answer']}")
    print()
```
